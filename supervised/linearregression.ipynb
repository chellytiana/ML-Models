# Linear Regression - Chapter 4
# Normal Equation, Predictions, and Scikit-Learn Implementation

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import add_dummy_feature
from sklearn.linear_model import LinearRegression
```

```python
np.random.seed(42)  # reproducibility

m = 100  # number of instances
X = 2 * np.random.rand(m, 1)
y = 4 + 3 * X + np.random.randn(m, 1)
```

```python
plt.figure(figsize=(8, 5))
plt.plot(X, y, "b.")
plt.xlabel("x1")
plt.ylabel("y")
plt.title("Randomly Generated Linear Dataset")
plt.grid(True)

plt.savefig("linear_regression_plot.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
X_b = add_dummy_feature(X)  # add bias term (x0 = 1)

theta_best = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ y

print("Theta (Normal Equation):")
print(theta_best)
```

```python
X_new = np.array([[0], [2]])
X_new_b = add_dummy_feature(X_new)

y_predict = X_new_b @ theta_best

print("Predictions:")
print(y_predict)
```

```python
plt.figure(figsize=(8, 5))
plt.plot(X_new, y_predict, "r-", linewidth=2, label="Predictions")
plt.plot(X, y, "b.")
plt.xlabel("x1")
plt.ylabel("y")
plt.legend()
plt.title("Linear Regression Predictions")
plt.grid(True)
plt.show()
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X, y)

print("Intercept:", lin_reg.intercept_)
print("Coefficient:", lin_reg.coef_)
```

```python
print("Sklearn Predictions:")
print(lin_reg.predict(X_new))
```
